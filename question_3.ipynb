```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read in the spotify and youtube data from the csv file
spotify_data = pd.read_csv('Spotify_Youtube.csv', index_col=0)
```

```python
#Look at the head of the dataframe to get a feel for the data
spotify_data.head()
```

```python
#Look at the distribution of the numerical variables in the dataframe
spotify_data.describe()
```

```python
#Universal function we used to determine if a song is 'most streamed'
#Returns 1 if the song is in the top 10% of streams, 0 otherwise
def top_10_percent(streams):
    threshold = spotify_data['Stream'].quantile(0.9)
    return 1 if streams >= threshold else 0
```

```python
#Apply the function to the dataframe to create a new column
spotify_data['Top 10%'] = spotify_data.apply(lambda row: top_10_percent(row['Stream']), axis=1)
```

```python
#Create a dataframe of the songs that are in the top 10% of streams
most_streamed = spotify_data[spotify_data['Top 10%'] == 1].copy()
```

```python
#Look at the distributions in this new dataframe
most_streamed.describe()
```

```python
#import needed libraries for sentiment analysis
import nltk
```

```python
#Download the sentiment analyzer that will be used and intializing a variable that llows access
#Create a function that will return the sentiment of a string passed
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()
def sentiment_scores(docx):
    return sent_analyzer.polarity_scores(docx.text)
```

```python
#import spacy and load the english model
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
#Now using the intialized variable, we can prep the sentiment analyzer for the dataframe
from spacy.tokens import Doc
Doc.set_extension("sentimenter",getter=sentiment_scores, force=True)
```

```python
#Access the compound score of the sentiment analyzer and create a column that holds the score
most_streamed['Track_sentiment'] = most_streamed['Track'].apply(lambda x: nlp(x)._.sentimenter['compound'])
```

```python
#Let's look at the distribution of the compound scores
most_streamed['Track_sentiment'].hist()
```

```python
#Visualizing the data in a similar way to the previous question but each indv point is visible
sns.scatterplot(x='Stream',y='Track_sentiment',data=most_streamed)
```

```python
#Now we are going to apply the same process to the original complete dataframe
spotify_data['Track_sentiment'] = spotify_data['Track'].apply(lambda x: nlp(x)._.sentimenter['compound'])
```

```python
#And compare the two using a kernel density plot
sns.kdeplot(x='Track_sentiment',data=spotify_data, hue='Top 10%')
```

```python
#Songs overall seemingly have neutral titles but is there a relationship?
#Import the needed libraries for the regression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#Witin the top ten most streamed, is there a relationship between sentiment and streams?
model0 = smf.ols("Stream ~ Track_sentiment", data=most_streamed).fit()
model0.summary()
```

The OLS regression shows there is not a relationship between streams and sentiment of track titles within the top 10%.

```python
#To actually answer our question, we need to look at the entire dataset
#This model will show us if there is a relationship between sentiment of track title
#and if a song is 'most streamed'
model2 = smf.ols("Q('Top 10%') ~ Track_sentiment", data=spotify_data).fit()
model2.summary()
```

Again, no relationship is shown between sentiment of track title and if a song is 'most streamed'. This concludes our analysis but- let's look at track titles with 0 sentiment scores and see what they look like:
